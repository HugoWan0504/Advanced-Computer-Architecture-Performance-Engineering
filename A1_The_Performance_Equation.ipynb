# Amdahl's Law

Recall from cs203 that Amdahl's Law limits the speed up an optimization can provide.  It's given as 

$$S_{tot} = \frac{1}{\left(\frac{f}{S}\right) + (1-f)}$$

Where $S$ is the speedup provided by the optimization, $f$ is the fraction of execution time affected by the optimization, and $S_{tot}$ is total speedup.
